```python
import pandas as pd
```

```python
first_column = []
for x in ["human_score", "normal_prompting", "CoT_prompting", "FCoT_prompting"]:
  for i in range(0,4):
    first_column.append(x)

second_column = []
for i in range(0,4):
  for x in ["msn", "onion", "cnn", "cnn_fact_check"]:
    second_column.append(x)

third_column = [3.5, 7.5, 2.25, 3.5, 1, 9, 2, 1, 1,1,9.5,1.5, 5, 2, 2, 2.33]
```

```python
df = pd.DataFrame()
df["scoring_techniques"] = first_column
df["article"] = second_column
df["score"] = third_column
df
```

```python
import seaborn as sns
```

```python
ax = sns.barplot(df, x = "article", y = "score", hue = "scoring_techniques")
sns.move_legend(
    ax, "lower center",
    bbox_to_anchor=(.5, 1), ncol=4, title=None, frameon=False,
)
```
