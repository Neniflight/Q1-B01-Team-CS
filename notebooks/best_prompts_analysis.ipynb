```python
import pandas as pd
```

```python
human_csv = pd.read_csv("data/human_res.csv")
human_csv
```

```python
human_data = human_csv[~human_csv.iloc[:, 1].isnull()]
human_data
```

```python
genai_csv = pd.read_csv("data/gen_ai_res.csv")
genai_csv.head()
```

```python
genai_data = genai_csv[~genai_csv.iloc[:, 3].isnull()]
genai_data.shape
# 3 different prompting types * 4 adjustment types = 12
```

```python
article_scores = human_data.iloc[:, 1:].mean()
article_scores
```

```python
article_scores.iloc[0]
```

```python
genai_data_scores_only = genai_data[["Prompting Type", "Adjustment Made", "Article 1 (msn) Score" , "Article 2 (onion) Score", "Article 3 (cnn) Score", "Article 4 (fact check) Score"]]
genai_data_scores_only
```

```python
def mae(row):
    sum_ae = 0
    for i in ["Article 1 (msn) Score", "Article 2 (onion) Score", "Article 3 (cnn) Score", "Article 4 (fact check) Score"]:
        prompt_score = row.loc[i] 
        article_score = article_scores.loc[i]
        ae = abs(prompt_score - article_score)
        sum_ae += ae
    mae = sum_ae / 4
    return mae
```

```python
genai_data_scores_only['MAE'] = genai_data_scores_only.apply(mae, axis=1)
genai_data_scores_only
```

```python
genai_data_scores_only.sort_values(by="MAE", ascending=True)
```

Normal Prompting with no adjustment seem to get the best results without any adjustments surprisingly. They seem to match what we think about naive realism. However, it could be because our sample is small, as I thought the model was thinking more with fractal chain of thought. Normal Prompting is the top 3 in MAE. 

```python
def mean_prompts(row):
    total_sum = 0
    for i in ["Article 1 (msn) Score", "Article 2 (onion) Score", "Article 3 (cnn) Score", "Article 4 (fact check) Score"]:
        prompt_score = row.loc[i] 
        total_sum += prompt_score
    mean_prompt = total_sum / 4
    return mean_prompt
```

```python
genai_data_scores_only['mean_of_prompts'] =  genai_data_scores_only.apply(mean_prompts, axis=1)
genai_data_scores_only
```

```python
genai_data_scores_only.sort_values(by="mean_of_prompts", ascending=False)
```
